```python
data = [
    'BE,AMS,BHX,E75,460',
    'BE,AMS,EMA,DH4,729',
    'BE,EDI,EXT,DH4,599',
    'BE,AMS,EXT,DH4,590',
    'BE,AMS,INV,DH4,409',
    'BE,AMS,SOU,E75,512',
    'BE,BRU,SOU,E82,111'
]
```

```python
def m(line):
    a,b,c,d,e = line.split(',')
    return [b,int(e)]
intermediate = list(map(lambda x:m(x), data))
```

```python
sorted(intermediate)
```

```python
def r(d):
    previous_airport = ''
    sum = 0
    count = 0
    for row in d:
        if row[0] == previous_airport:
            sum += row[1]
            count += 1
        else:
            if count > 0:
                print(previous_airport + ' ' + str(sum/count))
            previous_airport = row[0]
            sum = row[1]
            count = 1
    
r(sorted(intermediate))
```
